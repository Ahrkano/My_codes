```python
import pandas as pd
import numpy as np
import ast
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_data():


    file = open('myfile.txt', 'r')

    file = file.read()

    numbers = re.findall(r"[-+]?\d*\.\d+|\d+", file)

    return (numbers)
```

```python
def data_to_int(values):
    res = []
    for i in values:
        res.append (ast.literal_eval(i))
    return res
```

```python
teste = read_data()
teste_conv = data_to_int(teste)
teste_res = [teste_conv[n+1:n+4] for n in range(0, len(teste_conv), 4)]

Q = []
M = []
F = []

for i in teste_res :
    Q.append(i[0])
    M.append(i[1])
    F.append(i[2])
    
teste_pop = pd.DataFrame()
teste_pop["Population"] = Q
teste_pop["Male"] = M
teste_pop["Female"] = F
```

```python
size = len(teste_pop.index)
plt.figure(figsize=(16,8))
plt.plot( teste_pop)
plt.gca().legend(('População','Macho','Fêmea'), loc='upper left')
plt.xticks(np.arange(0, size+1, 35))
plt.title("Crescimento Populacional")
plt.xlabel('Dias passados')
plt.ylabel('Indivíduos')
plt.show()
```
