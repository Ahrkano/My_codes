```python
from PIL import Image
```

```python
imagem = Image.open('nic.bmp')
```

```python
pixels = list(imagem.getdata())
```

```python
def image_to_binary(aux):
    res = []
    for i in aux:
        R = bin(i[0])
        G = bin(i[1])
        B = bin(i[2])
        temp = (R,G,B)
        res.append(temp)
    return res
```

```python
def image_to_int(aux):
    res = []
    for i in aux:
        R = (int((i[0]), 2))
        G = (int((i[1]), 2))
        B = (int((i[2]), 2))
        temp = (R,G,B)
        res.append(temp)
    return res
```

```python
def message_to_binary(aux):
    res = []
    for i in aux:
        res.append(format(ord(i),  '09b'))
    return res
```

```python
def binary_to_message(aux):
    res = []
    for i in aux:
        res.append(chr(int(i, 2)))
    return res
```

```python
def esteg_message(picture, text):
    res = []
    bank = []
    x = 0
    y = 0
    
    for elem in text:
        for let in list(tuple(reversed(elem))):
            
            temp = (picture[x][y][:-1] + (let))
            bank.append(temp)
                        
            y += 1
            x += 1
            
            if y == 3 : 
                y = 0
    
    while (y != 0 and y < 3) :
        bank.append(picture[x][y])
        y+=1
        
    res =  [bank[n:n+3] for n in range(0, len(bank), 3)]
    
    bank.clear()
    for i in res:
        bank.append(tuple(i))
        
    return bank
```
